<b><span style="color: #4c1d95; background-color: #ede9fe; padding: 20px; font-size: 18px; border-left: 8px solid #a78bfa"> <strong>[1.4] üß™ Evaluation Metrics</strong></span></b>

Your model‚Äôs performance in this competition will be assessed using a **weighted multilabel Area Under the ROC Curve (AUC-ROC)** across 14 classification targets:

#### **üéØ Classification Targets**

1. **Aneurysm Present** ‚Äì a binary label indicating whether **any aneurysm is present** in the brain scan.
2. **13 Anatomical Locations** ‚Äì binary labels indicating the **presence or absence of aneurysms** in 13 specific brain artery locations (e.g., Left MCA, Basilar Tip, etc.).

---

#### **üî¢ Final Score Calculation**

The **final score** is a **weighted average** of the individual AUC-ROC scores across all 14 labels, giving **13√ó higher weight** to the critical `Aneurysm Present` label.

$$
\text{Final Score} = \frac{13 \cdot \text{AUC}_{\text{present}} + \sum_{i=1}^{13} \text{AUC}_i}{13 + 13}
$$

Or simplified:

$$
\text{Final Score} = \frac{13 \cdot \text{AUC}_{\text{present}} + \text{Sum of 13 other AUCs}}{26}
$$

---

#### **üìò Metric Details**

##### 1. **AUC-ROC (Area Under the Receiver Operating Characteristic Curve)**

The AUC-ROC measures your model‚Äôs ability to distinguish between positive and negative cases for each label:

* **AUC = 1.0** ‚Üí Perfect separation
* **AUC = 0.5** ‚Üí Random guessing
* **AUC < 0.5** ‚Üí Worse than random

For each label (e.g., ‚ÄúLeft MCA‚Äù), the AUC is computed between your predicted probability and the ground truth binary label (1 = aneurysm present, 0 = absent).

---

#### **‚öñÔ∏è Why Weight `Aneurysm Present` More?**

Detecting whether an aneurysm exists **anywhere** in the brain is the **most clinically critical** task‚Äîerrors here could mean a missed diagnosis entirely. By weighting it **13√ó more**, the scoring reflects this priority while still rewarding correct localization.

Women are more frequently affected than men, with peak incidence between ages 30 and 60. The most common form, a saccular (‚Äúberry‚Äù) aneurysm, looks like a small berry hanging from the artery by a narrow neck.

* **Post-rupture survival**: In untreated ruptured aneurysms, the median survival from symptom onset is \~20 days, and the one-year mortality rate is around 65 %.

* **Age**: Incidence peaks between ages 30 and 60, though aneurysms can develop at any adult age.
* **Risk factors**:
  * **Non-modifiable**: age, female sex, family history of aneurysm
  * **Modifiable**: hypertension, smoking, and certain connective-tissue disorders

#### ‚è© Types of Brain Aneurysms

* **Saccular (Berry) Aneurysms**
  The most common type (\~90%), these form a rounded pouch with a narrow neck protruding from one side of the artery wall. They typically occur at **arterial bifurcations** and are highly prone to rupture.

* **Fusiform Aneurysms**
  These involve circumferential dilation of the artery without a distinct neck, causing the entire segment to bulge. They‚Äôre less common and typically found in the **vertebrobasilar circulation**.

---

#### ‚è© Understanding the Brain‚Äôs Circulatory System

The brain is supplied by two major arterial systems:

* **Internal Carotid Arteries (ICA)**
  Arise from the neck and supply the **anterior circulation** (frontal, parietal lobes, and much of the cerebral cortex).
  ‚Üí Give rise to the **ACA**, **MCA**, and **PComA**.

* **Vertebral Arteries**
  Ascend alongside the spine and merge to form the **basilar artery**, supplying the **posterior circulation** (occipital lobes, brainstem, and cerebellum).
  ‚Üí Feed into the **PCA** and **brainstem branches**.

These arteries converge at the **Circle of Willis**, a ring-like arterial structure at the base of the brain that allows collateral blood flow.

> **Aneurysm hotspots**: Most brain aneurysms form at the **junctions** of arteries within or near the Circle of Willis ‚Äî especially at the **AComA**, **MCA bifurcation**, **ICA segments**, and **basilar tip** ‚Äî due to high hemodynamic stress and wall shear forces.

#### ‚è© Natural History & Critical Timeframes

**Sentinel headaches** often occur **6 to 20 days before a full aneurysm rupture**. These are sudden, severe headaches caused by minor blood leakage (a ‚Äúwarning leak‚Äù) from the aneurysm. While easily mistaken for migraines or tension headaches, they are a critical early indicator and should prompt immediate neuroimaging.

**Latency** refers to the unpredictable duration during which an unruptured intracranial aneurysm (UIA) remains stable. Some aneurysms may stay unchanged for years or even decades, while others may grow or rupture without warning. There is currently no reliable method to predict exactly when or if a UIA will rupture.

**The survival window after rupture** is extremely short without intervention. On average, patients survive about **20 days** post-rupture if untreated, and **approximately 65% die within the first year**. Rapid diagnosis and treatment are therefore vital, often within **the first 24‚Äì72 hours** to prevent rebleeding and long-term neurological damage.

```python
def dataframe_summary(df, name="DATAFRAME"):
    """
    Creates a summary of a DataFrame showing:
    - Missing values count and %
    - Number of unique values
    - Data types
    
    Parameters:
        df   : pandas DataFrame
        name : str, optional, a label for the dataset (default "DATAFRAME")
    
    Returns:
        pandas Styler object (summary table with gradient coloring)
    """
    columns = df.columns.tolist()
    
    # Missing value summary
    missing_summary = pd.DataFrame({
        'Feature': columns,
        f'[{name}] Missing Count': df.isnull().sum().values,
        f'[{name}] Missing %': (df.isnull().sum().values / len(df)) * 100
    })

    # Unique value summary
    unique_summary = pd.DataFrame({
        'Feature': columns,
        f'Unique Values [{name}]': df.nunique().values
    })

    # Data types
    dtypes_summary = pd.DataFrame({
        'Feature': columns,
        'Data Type': df.dtypes.values
    })

    # Merge all summaries
    summary = (
        missing_summary
        .merge(unique_summary, on='Feature', how='left')
        .merge(dtypes_summary, on='Feature', how='left')
    )

    # Return styled DataFrame
    return summary.fillna(0).style.background_gradient(cmap='viridis')
```

  &emsp;How can we interactively explore the full 3D volume of a DICOM series slice-by-slice?
</div>

```python
import os
import random
import numpy as np
import pydicom
import matplotlib.pyplot as plt
import base64
from io import BytesIO
from IPython.display import HTML

# --- Paths
dicom_dir = '/kaggle/input/rsna-intracranial-aneurysm-detection/series'

# --- Helper to load full DICOM volume
def load_dicom_volume(series_path):
    try:
        files = sorted(os.listdir(series_path), key=lambda f: pydicom.dcmread(os.path.join(series_path, f)).InstanceNumber)
        slices = []
        for f in files:
            dcm = pydicom.dcmread(os.path.join(series_path, f))
            img = dcm.pixel_array
            if img.ndim == 3:  # multi-frame
                img = img[0]
            slices.append(img)
        return np.stack(slices), dcm  # return last ds for metadata
    except Exception as e:
        print(f"Error loading series: {e}")
        return None, None

# --- Select one random valid series
valid_sids = [sid for sid in os.listdir(dicom_dir) if os.path.isdir(os.path.join(dicom_dir, sid))]
random.shuffle(valid_sids)

volume, meta = None, None
for sid in valid_sids:
    try:
        print(f"Trying series: {sid}")
        volume, meta = load_dicom_volume(os.path.join(dicom_dir, sid))
        if volume is not None and volume.shape[0] >= 5:  # filter for actual 3D volumes
            print(f"Successfully loaded series: {sid}")
            break
    except Exception as e:
        print(f"Error with series {sid}: {e}")
        continue

if volume is None:
    print("‚ö†Ô∏è No valid DICOM series found.")
else:
    # --- Metadata
    modality = getattr(meta, 'Modality', 'Unknown')
    age = getattr(meta, 'PatientAge', 'N/A')
    sex = getattr(meta, 'PatientSex', 'N/A')
    shape = f"{volume.shape[1]}√ó{volume.shape[2]}"
    num_slices = volume.shape[0]
    
    print(f"Loaded {modality} volume with {num_slices} slices")
    print(f"Patient Age: {age}, Sex: {sex}")
    
    # --- Normalize pixel values for better visualization
    volume_min = volume.min()
    volume_max = volume.max()
    if volume_max > volume_min:
        normalized_volume = (volume - volume_min) / (volume_max - volume_min)
    else:
        normalized_volume = volume
    
    # --- Create images for each slice and convert to base64
    # Limit to a reasonable number of slices to avoid browser performance issues
    max_slices = min(num_slices, 100)  # Limit to 100 slices max
    step = max(1, num_slices // max_slices)
    
    base64_images = []
    for i in range(0, num_slices, step):
        plt.figure(figsize=(8, 8), dpi=80)
        plt.imshow(normalized_volume[i], cmap='gray')
        plt.axis('off')
        plt.title(f"{modality} | Age: {age} | Sex: {sex} | Shape: {shape} | Slice {i+1}/{num_slices}", 
                 fontsize=12, fontweight='bold')
        plt.tight_layout()
        
        # Convert plot to base64 string
        buffer = BytesIO()
        plt.savefig(buffer, format='png', bbox_inches='tight')
        buffer.seek(0)
        img_str = base64.b64encode(buffer.read()).decode('utf-8')
        base64_images.append(img_str)
        plt.close()
    
    # --- Create HTML with embedded JavaScript slider
    html_content = f"""
    <div style="text-align:center; max-width:800px; margin:0 auto;">
        <div style="margin-bottom:10px;">
            <img id="sliceImage" src="data:image/png;base64,{base64_images[0]}" 
                 style="max-width:100%; height:auto; border:1px solid #ddd;">
        </div>
        <div style="margin:10px 0;">
            <input type="range" id="sliceSlider" min="0" max="{len(base64_images)-1}" value="0" 
                   style="width:80%; margin:0 auto;">
            <div id="sliceLabel" style="margin-top:5px; font-weight:bold;">
                Slice 1/{len(base64_images)}
            </div>
        </div>
        <div style="margin:10px 0;">
            <button id="playButton" style="padding:5px 15px;">‚ñ∂Ô∏è Play</button>
            <button id="stopButton" style="padding:5px 15px; margin-left:10px;">‚èπÔ∏è Stop</button>
        </div>
    </div>
    
    <script>
        // Store images in a JavaScript array
        const images = [];
        {' '.join([f"images.push('data:image/png;base64,{img}');" for img in base64_images])}
        
        // Get elements
        const slider = document.getElementById('sliceSlider');
        const image = document.getElementById('sliceImage');
        const label = document.getElementById('sliceLabel');
        const playButton = document.getElementById('playButton');
        const stopButton = document.getElementById('stopButton');
        
        // Update image when slider changes
        slider.oninput = function() {{
            const index = parseInt(this.value);
            image.src = images[index];
            label.textContent = `Slice ${{index*{step}+1}}/${{images.length*{step}}}`;
        }};
        
        // Animation variables
        let animationId = null;
        let currentIndex = 0;
        
        // Play function
        function playSlides() {{
            if (animationId) return; // Already playing
            
            animationId = setInterval(() => {{
                currentIndex = (currentIndex + 1) % images.length;
                slider.value = currentIndex;
                image.src = images[currentIndex];
                label.textContent = `Slice ${{currentIndex*{step}+1}}/${{images.length*{step}}}`;
            }}, 200); // Change slice every 200ms
        }}
        
        // Stop function
        function stopSlides() {{
            if (animationId) {{
                clearInterval(animationId);
                animationId = null;
            }}
        }}
        
        // Add event listeners
        playButton.onclick = playSlides;
        stopButton.onclick = stopSlides;
    </script>
    """
    
    # Display the HTML
    display(HTML(html_content))
```
